# **Variáveis**
Variáveis são containers para armazenar valores de dados;
uma variável é criada quando você atribui um valor a ela


```python
#Variáveis de Texto

Nome = 'Abner Lima'
var_texto = str('1A2B3')
print(Nome, '\n',var_texto)
```

```python
#Variáveis Numéricas

var_01 = 100
var_02 = int(200)
var_03 = float(10.5)
var_04_booleano = True
var_05_booleano = False

print(var_01, var_02, var_03, var_04_booleano, var_05_booleano)
```

# **Tipos de Dados**

*   listas- []
*   Tuplas- ()
*   Dicionários- {}
**OBS:** Esses tipos de dados pode receber qualquer tipo de informação, inclusive Operações Aritméticas






```python
#Listas

Lista_Exemplo_01 = [1, 2, 3, 4]
Lista_Exemplo_02 = ['Nome', True, 1, [2, 3, 4]]

print(Lista_Exemplo_01, Lista_Exemplo_02)
```

```python
#Tuplas !!!!!!! SÃO IMUTAVEIS, criada uma vez, não pode ser alterada!

Tupla_01 = (1, 2, 3, 4)
Tupla_02 = ('Abner', True, 1.65)

print(Tupla_01, Tupla_02)
```

```python
#Dicionário

Dicionario = {
    'index': var_texto,
    'id': var_01,
    'nome': Nome,
    'lista': Lista_Exemplo_01,
    'tupla': Tupla_02
    }
Dicionario
```

# **Nomeação de Variáveis**




```python
#Declarar variáveis em uma única linha de comando

Laranja, Uva, Morango = 1, 2, 3
print(Laranja, Uva, Morango)
```

```python
#Declarar variáveis atribuindo um valor a várias variáveis

Laranja = Morango = Uva = 1
print(Laranja, Morango, Uva)
```

```python
#Nomear com listas

Carros = ['BMW', 'MERCEDES', 'AUDI']
Marca_01, Marca_02, Marca_03 = Carros
print(Marca_01, Marca_02, Marca_03)
```

```python
#Combinar variáveis

Nome1= 'Abner'
Sobrenome = 'Lima'
Nome_Completo = Nome1 + ' ' + Sobrenome
print(Nome_Completo)
```

```python
#Exemplo

var1= 'Fabiana'
var2 = str(36)
var1 + ' tem ' + var2 + ' Anos'
```
